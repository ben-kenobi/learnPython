---
背景：

* 你现在负责公司的产品战略规划
* 你带领团队分析各大洲的人口变化及分布等数据
* 要要将人口分布数据按大洲切分成若干个 Excel

---
解题步骤：

* 世界人口数据在 gap 中
* 取得所有不重复的大洲，放到 continents 中
* 循环遍历 continents，有一个变量存储每一次循环的 continent

````python
# 非题目，取得世界人口相关数据
gap = px.data.gapminder()
gap
````

````python
#* 取得所有不重复的大洲，放到 continents 中
continents = gap.continent.unique()
continents
````

````python
"""
* 循环遍历 continents，有一个变量存储每一次循环的 continent
      用 continent 从 gap 取这个大洲的数据，放到 df 
      用 continent 拼接出要保存的文件名 file_name
      把 df 里的数据保存到 file_name 文件中
"""
from pathlib import Path
for cont in continents:
    df = gap[gap.continent == cont]
    file_name = f'{cont}.xlsx'
    df.to_excel(file_name,index=False)
````

# 切分各大洲/国家的人口数据
---
备注: 这里综合了 p187 和之前数据过滤的内容

* 使用 unique() 函数得到某一列不重复值
* 这里又超纲学习了 pathlib 来创建目录
```
  Path("父目录/子目录").mkdir(parents=True, exist_ok=True)    
  Path("BG1/BU1").mkdir(parents=True, exist_ok=True)    
```  
  Path() 里的参数是目录路径，可以只有一层 "目录1", 
  
  也可以有N层: "BG1/BU1/SKU1", "爷爷目录/爸爸目录/儿子目录/孙子目录"
  
  parents=True  表示当 "爷爷目录/爸爸目录/儿子目录", 有任何一层目录不存时，它会自动帮你补创建好
  
  exist_ok=True 表示如果创建这个目录已存时，它会自动跳过，不报错
  

* [pathlib 介绍](https://juejin.im/post/6844903721017540621)
* [pathlib 官方文档](https://docs.python.org/zh-cn/3/library/pathlib.html)
---
背景：

* 接上一题
* 公司发现人口信息要深挖的信息太多了
* 要求你把为每个大洲建一个目录
* 然后每个国家一个 xlsx 文件，放到国家所在的大洲目录里

---
解题步骤：

* 对 gap 的 ['continent','country'] 进行 groupby + sum, 结果放到 stats

   sum/count 都可以，主要目的是为取出不重复的 ['continent','country'] 组合出来
   
* 取出 stats 的索引，放到 continent_countries

* 循环遍历 continent_countries，有一个变量存储每一次循环的 continent_country
      从 continent_country 中取得大洲名 continent
      从 continent_country 中取得国家名 country
      用大洲名(continent) 创建一个目录（如果存在也不报错exist_ok=True）
         Path(continent).mkdir(parents=True, exist_ok=True)    
      用 continent,country 拼接出要保存的文件名 file_name
         因为是要保存在 continent 目录下，所以在continent,country之间要有"/"
      用 continent,country 从 gap 取这个大洲/国家的数据，放到 df 
      把 df 里的数据保存到 file_name


---


````python
# 对 gap 的 ['continent','country'] 进行 groupby + sum, 结果放到 stats
#  sum/count 都可以，主要目的是为取出不重复的 ['continent','country'] 组合出来
status = gap.groupby(['continent','country']).sum()
status
````

````python
# 取出 stats 的索引，放到 continent_countries
continent_countries = status.index
continent_countries
````

````python
# 导入 pathlib 里相关的包，(非题目)
from pathlib import Path
````

````python
"""
* 循环遍历 continent_countries，有一个变量存储每一次循环的 continent_country
      从 continent_country 中取得大洲名 continent
      从 continent_country 中取得国家名 country
      用大洲名(continent) 创建一个目录（如果存在也不报错exist_ok=True）
         Path(continent).mkdir(parents=True, exist_ok=True)    
      用 continent,country 拼接出要保存的文件名 file_name
         因为是要保存在 continent 目录下，所以在continent,country之间要有"/"
      用 continent,country 从 gap 取这个大洲/国家的数据，放到 df 
      把 df 里的数据保存到 file_name
"""

for con,country in continent_countries:
    file_p = f'./{con}/'
    Path(file_p).mkdir(parents=True,exist_ok=True)
    file_n = f'{file_p}{country}.xlsx'
    df = gap[(gap.continent == con) & (gap.country == country)]
    df.to_excel(file_n)
````

````python
# 过滤 gap 中 year == 2007 的数据 (使用 df.query)
# 对大洲(continent)升序，人口(pop)降序进行排序  (df.sort_values)
# 将数据放到 df_2007 中
# by=['continent','pop'], 表示同时对这两个字段进行排序
# ascending=[True,False]，表示对第1个字段(continent)进行升序，第2个字段(pop)进行降序排序
df_2007 = gap.query('year == 2007')
df_2007 = df.sort_values(by=['continent','pop'],ascending=[True,False])
df_2007 = df_2007[['continent','country','pop','lifeExp']]
df_2007
````

````python
# describe函数同时输出 几个 聚合函数 的结果
gap.groupby('continent').lifeExp.describe()
````

````python
# 自定义聚合函数
def my_mean(values):
    return values.mean()
gap.groupby('continent').lifeExp.agg(my_mean)
````

````python
# 自定义带额外参数的聚合函数
def my_mean_diff(values,diff_value):
    return values.mean() - diff_value
global_lifexp_mean = gap.lifeExp.mean()
gap.groupby('continent').lifeExp.agg(my_mean_diff,diff_value = global_lifexp_mean)
````

````python
# 同时传递多个聚合函数
gap.groupby('continent').lifeExp.agg([np.sum,my_mean])
````

````python
# 使用字典调用agg
gap.groupby('continent').agg({'lifeExp' : np.mean , 'pop' : np.sum})
````

````python
# 转换函数，将数据转换后返回
def year_2_month(year):
    return year * 12;
print(gap.lifeExp.transform(year_2_month))
gap.groupby('continent').lifeExp.transform(year_2_month)
````

````python
# filter 函数 对分组后的数据进行 聚合过滤
gap_gt_50 = gap.groupby('continent').filter(lambda ser: ser.lifeExp.mean() > 60)
gap_gt_50
````

````python
# 使用for 循环遍历分组数据
grouped = gap.groupby('continent')
# print(grouped.groups)
print(grouped.get_group('Africa'))
for g_df in grouped:
    print(type(g_df))
````

````python
# 多级索引的分组数据，对分组后的数据再执行分组操作
coun_ser = gap.groupby(['continent','country']).lifeExp.count()
print(coun_ser)
coun_ser.groupby(level=[0,1]).count()
````

````python
# 对 df_2007 中有数据按大洲(continent)进行分组(groupby)
# 并按人员数量的降序(ascending = False)，对国家进行排名
# 将排名信息放到 df_2007 的 rank 列中
rk = df_2007.groupby('continent').pop.rank(ascending=False)
rk
df_2007['rank'] = rk
df_2007
````

````python
# 查看亚洲(Asia)的排名数据, (使用 df.query 函数)
df_asia_rk = df_2007.query('continent == "Asia"')
df_asia_rk
````

````python
# 从 df_2007 里查询所有 rank 小于4 的数据，放到every_continet_top3
#  使用 df.query 来查询，如果大家不习惯 query, 还是可以用之前的 [] 来查询数据
every_continet_top3 = df_2007.query('rank < 4')
every_continet_top3
````

````python
# 对df_2007 的数据按 continent 的分组(groupby)
# 再对人口(pop)字段进行累计求和(累加 cumsum)
# 把累加结果放到 df_2007 里的 cumsum 列
pop_sum = df_2007.groupby(['continent']).pop.cumsum()
df_2007['cumsum'] = pop_sum
df_2007
````

````python
#查看一下亚洲的数据（非题目）
asia_df = df_2007.query('continent == "Asia"')
````

````python
# 画一下亚洲人口的累计分布曲线，从图可以看出
# 中国和印度的人口占了超过亚洲一半 （非题目)
asia_df = asia_df[['country','cumsum']]
asia_df = asia_df.set_index('country')
asia_fig = asia_df.plot()
asia_fig.show()
````

````python
#为了能在 GitHub 中也能看到图，我使用了 show('png'), (非题目)
asia_fig.show('png')
````

````python
# 使用 px.treemap 画出 2007年世界人口各国人口占比图
# px.treemap 函数各个参数的介绍
#   df_2007 是它的数据参数
#   path=['continent', 'country'], 表名图表的显示顺序
#        先显示大洲(continent), 国家/地区(country)
#   values = 'pop' 表示树型图以人口(pop)为数值（方块大小）
#   color = 'lifeExp' 表示树型图以人口预计寿命(lifeExp)为方块颜色
fig = px.treemap(df_2007,path=['continent', 'country'],values = 'pop',color='lifeExp')
fig.show()
````

---
备注: 这里综合了 p176 的内容

因为前面的题目都比较难，这一题的内容较为简单，让大家快乐的结束这一章

---
背景：

* 公司想看每个大洲的各个国家人口，最多是多少人，最少是多少人，总和是多少

---
解题步骤：

* 对gap 的大洲(continent) 进行分组 groupby
  然后调用聚和(agg)函数，传入参数: ["max","min","sum"]
  将值赋给 continent_stat

* 使用 reset_index 函数，将continent_stat里的索引变成列；
  通过重置索引，将分层的结果平铺展示



````python
"""
对gap 的大洲(continent) 进行分组 groupby
  然后调用聚和(agg)函数，传入参数: ["max","min","sum"]
  将值赋给 continent_stat
"""
stat = gap.groupby('continent')['pop'].agg([np.max,np.min,np.sum])
stat
````
